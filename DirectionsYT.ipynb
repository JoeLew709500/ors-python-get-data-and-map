```python
import openrouteservice as ors
import folium
import operator
import json
import os
from functools import reduce
from folium.plugins import FloatImage
import re
```

```python
client = ors.Client(base_url='http://localhost:8080/ors')
```

```python
coords = 
routeName = input('Enter Route Name')

# Set the initial location to the first coordinate
initial_location = [coords[0][1], coords[0][0]]

# Create a map
mapio = folium.Map(location=initial_location, zoom_start=15)

# Get the route from the client
route = client.directions(coordinates=coords, profile='driving-car', format='geojson')

# Initialize previous instruction
previous_instruction = None

# Get the instructions and corresponding waypoints from the route, skipping those with "Arrive" in the instruction or if the previous instruction is the same
instructions = []
waypoints = []
for segment in route['features'][0]['properties']['segments']:
    for step in segment['steps']:
        if not step['instruction'].startswith('Arrive') and step['instruction'] != previous_instruction:
            instructions.append(step['instruction'])
            waypoints.append(step['way_points'][0])  # Assuming each step has at least one waypoint
            previous_instruction = step['instruction']

# Add the route to the map
folium.PolyLine(locations=[list(reversed(coords)) for coords in route['features'][0]['geometry']['coordinates']], color='blue').add_to(mapio)

# Initialize RGB values
r, g, b = 0, 255, 0

# Add markers to the map
for i, index in enumerate(waypoints):
    coordinates = list(reversed(route['features'][0]['geometry']['coordinates'][index]))
    instruction = instructions[i]
    tooltip_text = f"{i}: {instruction}"
    
    # Adjust color every 10 waypoints
    if i % 10 == 0 and i != 0:
        g = max(0, g - 5)
        r = min(255, r + 5)
        b = max(0, b - 10)
    
    color = f"#{r:02x}{g:02x}{b:02x}"
    radius = 2
    weight = 3
    
    if i == 0:
        color = 'purple'
        radius = 5
        weight = 5
    elif i == len(waypoints) - 1:
        color = 'black'
        radius = 5
        weight = 5
    
    circle = folium.CircleMarker(location=coordinates, radius=radius, color=color,
                                 fill=False, fill_color=color, tooltip=tooltip_text, weight=weight).add_to(mapio)
    
# add compass to map
compass_image =  (
    "https://raw.githubusercontent.com/ocefpaf/secoora_assets_map/"
    "a250729bbcf2ddd12f46912d36c33f7539131bec/secoora_icons/rose.png"
)

FloatImage(compass_image, bottom=5, left=5).add_to(mapio)

# Define the folder path
folder_path = r"C:\Users\joele\OneDrive\Desktop\pythonrouting"

# Create the folder if it doesn't exist
if not os.path.exists(os.path.join(folder_path, routeName)):
    os.makedirs(os.path.join(folder_path, routeName))

# Define the file paths
file_path = os.path.join(folder_path, f"{routeName}\{routeName}_route_data.json")
map_path = os.path.join(folder_path, f"{routeName}\{routeName}_route_map.html")
instructions_path = os.path.join(folder_path, f"{routeName}\{routeName}_instructions.txt")

# Save the instructions to a text file
with open(instructions_path, 'w') as file:
    for index, instruction in enumerate(instructions):
        file.write(f"{index + 1}. {instruction}\n")

# Save the route data to a json file
with open(file_path, 'w') as json_file:
    json.dump(route, json_file, indent=4)

# Save the map to an html file
mapio.save(map_path)

num_waypoints = len(waypoints)
num_instructions = len(instructions)

print(f"waypoints: {num_waypoints}, instructions: {num_instructions}")
```
